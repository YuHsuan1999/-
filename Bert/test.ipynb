```python
import pymysql
db_settings = {
    "host": "127.0.0.1",
    "port": 3306,
    "user": "root",
    "password": "",
    "db": "news",
    "charset": "utf8"
}

try:
    conn = pymysql.connect(**db_settings)
except Exception as ex:
    print(ex)
```

```python
import pandas as pd
# 確認是否已在DB中

with conn.cursor() as cursor:
    sql = 'SELECT Comment FROM comments WHERE Provenance = \'三立\''
    cursor.execute(sql)
    result = cursor.fetchall()

    count = 0
    text = []

    for row in result:
        text.append(row[0])
        print(row[0])
        count += 1

print('--------------------------\n' + str(count))
```

```python
df = pd.DataFrame(text)
df
```
