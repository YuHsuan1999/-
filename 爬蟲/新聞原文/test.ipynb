```python
import time
import datetime
from apscheduler.schedulers.background import BackgroundScheduler
from apscheduler.triggers.interval import IntervalTrigger
import news

def my_job():
    news.SETN()
    news.EBC()
    news.TVBS()

if __name__ == "__main__":
    scheduler = BackgroundScheduler()
    intervalTrigger = IntervalTrigger(hours=12)
    scheduler.add_job(my_job, intervalTrigger, id='news')

    scheduler.start()
```
