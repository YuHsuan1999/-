```python
import pymysql
db_settings = {
    "host": "127.0.0.1",
    "port": 3306,
    "user": "root",
    "password": "",
    "db": "news",
    "charset": "utf8"
}

try:
    conn = pymysql.connect(**db_settings)
except Exception as ex:
    print(ex)
```

```python
import pandas as pd
# 確認是否已在DB中

with conn.cursor() as cursor:
    sql = 'SELECT News_id, Title, Provenance FROM original'
    cursor.execute(sql)
    result = cursor.fetchall()

    for row in result:
        print(row[1])
        with conn.cursor() as cursor:
            # 新增資料SQL語法
            keywords = 'INSERT INTO keywords VALUES(%s, %s, %s)'
            cursor.execute(keywords, (row[0], row[2], '1'))
         # 儲存變更
        conn.commit()
```
