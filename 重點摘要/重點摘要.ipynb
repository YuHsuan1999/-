```python
# spaCy 載入中文模型
import spacy
import zh_core_web_lg
import neuralcoref

nlp = zh_core_web_lg.load()
neuralcoref.add_to_pipe(nlp)

# summarizer 載入中文模型
from summarizer import Summarizer
from summarizer.sentence_handler import SentenceHandler
from spacy.lang.zh import Chinese
from transformers import *

# Load model, model config and tokenizer via Transformers
modelName = "bert-base-chinese" # 可以換成自己常用的
custom_config = GPT2Config.from_pretrained('gpt2')
custom_config.output_hidden_states = True
custom_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
custom_model = GPT2Model.from_pretrained('gpt2', config=custom_config)
custom_model = AutoModel.from_pretrained(modelName, config=custom_config)

model = Summarizer(
    custom_model=custom_model, 
    custom_tokenizer=custom_tokenizer,
    sentence_handler = SentenceHandler(language=Chinese)
    )

```

```python
body = "本月2日，台鐵太魯閣號發生出軌事故，肇因為工程車滑落邊坡，導致列車衝撞釀禍，目前工程車駕駛、義祥負責人李義祥，以及當天也在場的越南籍逃逸移工「阿好」，已雙雙被法院裁定收押，而2人對案發時的說詞卻有些微出入，「阿好」供稱，當時應固定在工程車上的綁帶，其實根本沒綁好，是脫落而非被扯斷。據了解，雖然李義祥跟「阿好」均稱，當時是工程車過彎不慎滑落，2人企圖以怪手拉起、卻失敗釀禍，不過李義祥稱綁帶斷裂，導致工程車掉落鐵軌，但「阿好」卻透露，本應固定在工程車上的綁帶，其實「根本沒綁好」，所以瞬間脫落、而不是被扯斷。檢方查扣綁帶，發現有多處裂痕，但究竟是被拉斷還是掉落尚待釐清，已將其送驗。由於事故釀成重大傷亡，為防止李義祥脫產，台鐵自本月4日起，向法院聲請對李義祥、義祥工業社、東新營造有限公司聲請假扣押，隨後，法院查封了3.7億餘元，7日，台鐵再拿出4.4億餘元假扣押裁准，並向法院聲請強制執行，累計目前已查封約8.2億餘元。"

result = model(body)
full = ''. join(result)
print(full) # 摘要出來的句子
```
